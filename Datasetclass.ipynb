```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats as sp

class Dataset:

    def __init__(self, data, name="Dataset"):
        self.data = np.array(data)
        self.name = name
        self.is_loaded = True

    def __repr__(self):
        return (f"Dataset(name='{self.name}', "
                f"shape={self.data.shape}, dtype={self.data.dtype})")

    def load(self, data, name=None):
        if isinstance(data, str):
            import pandas as pd
            self.data = pd.read_csv(data).values
            print(f"Loaded from file: {data}")
        elif data is not None:
            self.data = np.array(data)
            print(f"Loaded! Shape: {self.data.shape}")
        else:
            print("No data provided.")
            return
        if name:
            self.name = name
        self.is_loaded = True

    def describe(self):
        data = self.data.flatten()
        result = {
            "count"  : len(data),
            "mean"   : np.mean(data),
            "std"    : np.std(data),
            "min"    : np.min(data),
            "max"    : np.max(data),
            "median" : np.median(data),
            "mode"   : sp.mode(data).mode[0],
        }
        for k, v in result.items():
            print(f"  {k:<8}: {v:.4f}" if isinstance(v, float)
                  else f"  {k:<8}: {v}")
        return result

    def clean(self, strategy="mean", z_thresh=3.0):
        cleaned = self.data.astype(float).copy()
        nan_count = np.sum(np.isnan(cleaned))
        if nan_count > 0:
            fill = (np.nanmean(cleaned) if strategy == "mean"
                    else np.nanmedian(cleaned))
            cleaned = np.where(np.isnan(cleaned), fill, cleaned)
            print(f"Filled {nan_count} NaN(s) using '{strategy}'")
        z = np.abs((cleaned - np.mean(cleaned)) / np.std(cleaned))
        cleaned = cleaned[z <= z_thresh]
        self.data = cleaned
        return cleaned

    def export(self, filepath, fmt="csv"):
        if fmt == "csv":
            np.savetxt(filepath, self.data, delimiter=",")
        elif fmt == "npy":
            np.save(filepath, self.data)
        print(f"Exported to {filepath}")

    def visualize(self, kind="hist"):
        fig, ax = plt.subplots(figsize=(8, 4))
        if kind == "hist":
            ax.hist(self.data.flatten(), bins=20,
                    color="#1D9E75", edgecolor="white")
        elif kind == "box":
            ax.boxplot(self.data.flatten(), patch_artist=True)
        ax.set_title(f"{self.name} — {kind}")
        plt.tight_layout()
        plt.show()
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")

    def __repr__(self):
        return f"BankAccount(owner='{self.owner}', balance={self.balance})"


acc = BankAccount("Willis", 1000)
acc.deposit(500)
acc.withdraw(200)
print(acc)
# >> BankAccount(owner='Willis', balance=1300)
        
```

```python
def find_second_largest(numbers):
    unique_numbers = sorted(list(set(numbers)))

    if len(unique_numbers) < 2 :
        return 'None'
    return unique_numbers[-2]

print(find_second_largest([10, 5, 10, 8, 3]))  # >> 8
print(find_second_largest([5, 5, 5]))           # >> None
print(find_second_largest([1, 2]))              # >> 1    
```

```python
#Check for prime number
import math
def is_prime(n):
    if n<= 0:
        return False
    if n == 2:
        return True    
    if n % 2 == 0:
        return False
    limit = int(math.sqrt(n)) +1
    for i in range(3, limit, 2):
        if n% i == 0:
            return False
    return True

sample = 21
print(is_prime(sample))
        
```

```python
#Moving average function
import numpy as np
def moving_average(arr, k):
    arr = np.array(arr)
    result = print([np.mean(arr[i:i+k]) for i in range(len(arr)-k+1)])
    return result


arr = np.array([1, 2, 3, 4, 5, 6, 7])
moving_average(arr, k=3)
# >> [2. 3. 4. 5. 6.]    
```

```python
## Implementing Stack data structure in Python

class Stack:
    def __init__(self):
        self.items =[]
        
    def is_empty(self):
        return len(self.items) == 0
    def push(self, item):
        self.items.append(item)
    def pop(self):
        if self.is_empty():
            print('The stack is empty!')
            return None
        return self.items.pop()    

    def peek(self):
        if self.is_empty():
            print("Stack is empty")
            return None
        return self.items[-1]

    def __repr__(self):
        return f"Stack({self.items})"

        
```

```python
s = Stack()
s.push(1)
s.push(2)
s.push(3)
print(s.peek())   # >> 3
print(s.pop())    # >> 3
print(s)          # 
```

```python
#Anagram checker function

def anagram_checker(str1, str2):
    cleaned_str1 = str1.replace(" ", "").lower()
    cleaned_str2 = str2.replace(" ", "").lower()
    #Check the length of string 1 and 2, if not equal, they aren't anagram:
    if len(cleaned_str1) != len(cleaned_str2):
        return False
    return sorted(cleaned_str1) == sorted(cleaned_str2)




print(anagram_checker("silent", "listen"))    
```

```python
#Function to group by frequency, returning a dictionary

from collections import Counter
def group_by_frequency(lst):
    counts = Counter(lst)

    grouped = {}
    for item, freq in counts.items():
        if freq not in grouped:
            grouped[freq] =[]
        grouped[freq].append(item)

    return grouped    
```

```python
data = ['a','a','b','c','a', 'b']
group_by_frequency(data)
```

```python
#Cosine Similarity function
import  numpy as np
def cosine_similarity(a,b):
    a = np.array(a)
    b = np.array(b)
    magnitude = np.linalg.norm(a) * np.linalg.norm(b)
    dot_product= np.dot(a,b)
    result = dot_product/magnitude

    return result

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
cosine_similarity(a, b)  # >> 0.9746    
```

```python
#Binary Search

def binary_search(arr, target):
    low = 0
    high = len(arr)-1
    mid = (low + high)//2
    guess = arr[mid]

    while low <= high:
        if guess == target:
            return mid
        if guess < target:
            high = mid - 1
        else:
            guess > target
            low = mid + 1

    return -1


print(binary_search([1, 3, 5, 7, 9, 11, 13], 7))   # >> 3

    
```

```python
print(binary_search([1, 3, 5, 7, 9, 11, 13], 6))   # >> -1
```

```python
print(binary_search([1, 3, 5, 7, 9, 11, 13], 13))  # >> 6
```

```python
#Implementing Queue data structures

class Queue:
    def __init__(self):
        self.items =[]
    def enqueue(self, item):
        self.items.append(item)
    def dequeue(self):
        if not self.items:
            print("Queue is empty")
            return None
        return self.items.pop(0)
           

    def peek(self):
        if not self.items:
            print("Queue is empty")
         return self.items[0]
    def size(self):
         return len(self.items)
    def __repr__(self):
         return f"Queue({self.items})"
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.peek())     # >> 1
print(q.dequeue())  # >> 1
print(q.size())     # >> 2
print(q)            # >> Queue([2, 3])         
```
