# SEABORN TUTORIAL


Seaborn is a python visualization library built on top of  Matplotlib, another python visualization library

It enables us to easily create the most common types of plots

It is easy to use, i.e It handles alot of complexities behind the scenes. It works well with pandas DataFrames

```python
#Getting started

import seaborn as sns
import matplotlib.pyplot as plt

height  =[62,64,69,75,66,68,65,71,76,73]
weight = [120, 136,148,175, 137,165,154,172,200,187]

sns.scatterplot(x= height, y= weight)
plt.show()
```

```python
gender = ['female', 'female', 'female','male', 'male','male','female','male', 'male', 'male']
sns.countplot(x= gender)
plt.show()
```

```python
# USe inbult seaborn tips dataset

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
# Include hue
sns.scatterplot(x= 'total_bill', y= 'tip', data = tips, hue = 'sex')
plt.show()
```

# STATISTICAL ESTIMATION PLOTS

Use relplot() seaborn function to plot statistical relationships between data variables

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(data = tips, x='total_bill', y  = 'tip',hue ='smoker', style = 'smoker', size = 'size')
plt.show()
```

```python
# Use relplot() to plot a line plot
dots = sns.load_dataset('dots')
dots.head()
```

```python
sns.relplot(data = dots, kind = 'line', x = 'time', y = 'firing_rate',hue = 'choice', size = 'coherence',col = 'align', style ='choice')
plt.show()
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(data = fmri, kind = 'line', x = 'timepoint', y= 'signal', col = 'region', hue ='event', )
```

# Distributional Representation

Use the displot() to visualise distributed data

```python
sns.displot(data = tips, x =  'total_bill', kde = True, col = 'time')
```

```python
sns.displot(data = tips, x =  'total_bill', col = 'time', hue = 'smoker', rug =True, kind = 'ecdf')
```

# Plots for Categorical Data

Use the seaborn catplot() function to plot a variety of categorical data plots

```python
sns.catplot(data = tips, x ='day', y ='total_bill',kind = 'swarm',hue = 'smoker')
```

```python
sns.catplot(data = tips, x ='day', y ='total_bill',kind ='violin')
```

# Multivariate view of complex dataset

Some seaborn functions combine multiple kinds of plot to quickly give us information summaries of a dataset
Examples of these functions include:
1. pairplot()
2. jointplot()

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.jointplot(data = penguins, x='flipper_length_mm', y = 'bill_length_mm', hue = 'species')
plt.show()
```

```python
sns.pairplot(data = penguins, hue = 'species')
```
