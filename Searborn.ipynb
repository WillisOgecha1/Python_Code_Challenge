# Relating variables with Scatter Plots

A scatter plot is a mainstay statistical visualization that depicts the joint distribution of two variables, using a cloud of points, where each point represents an observation. 
Lets plot some scatter plots


```python
tips= sns.load_dataset('tips')
sns.relplot(x='total_bill', y = 'tip', data =tips )
plt.show()
```

```python
# Using a third variable to color the points.
# This is achieves using hue semantic, which makes the colors gain meaning in the plot
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'smoker')
plt.show()
```

```python
# Use a different maker style to emphasize the difference between the classes, and to improve accessibility.
sns.relplot(data = tips, x ='total_bill', y = 'tip', hue = 'smoker', style ='smoker')
plt.show()
```

Representing 4 variables by assigning the hue and style semantics different variables


```python
sns.relplot(data = tips, x= 'total_bill', y = 'tip', hue = 'smoker', style ='time')
plt.show()
```

```python
#Qualitative palette vs Sequential palette
sns.relplot(data = tips, x= 'total_bill', y ='tip', hue ='size')
plt.show()
```

* Note: when the hue semantic is a categorical variable, it will use a default qualitative color palette, whereas if it is  numerical variable, it will use a sequential color palette as illustrated above

```python
# Using the third semantic : Size

sns.relplot(data = tips, x= 'total_bill', y ='tip', hue = 'smoker', size = 'size')
plt.show()
```

# Visualizing Distribution of data

The most common approach to visualising a distribution is the histogram.
We use displot() function to display the distribution of data. displot() is a figure-level function. a high level API from wich you can switch between the axes-level plots it holds i.e histogram, kde and ecdf

Lets go

```python
penguins
```

```python
sns.displot(penguins, x = 'flipper_length_mm')
plt.show()
```

```python
# Choose bi size
sns.displot(penguins, x = 'flipper_length_mm', binwidth = 3)
plt.show()
```

```python
#Conditioning on other vaariables

sns.displot(penguins, x = 'flipper_length_mm', hue = 'species')
plt.show()
```

By default, the differnt hstograms are layered on top of each other , ojn some cases, they may be difficult to distinguish, to solve this issue, we can employ some options.

```python
# 1. Change visual representation of the histogram to a 'step' plot
sns.displot(penguins, element = 'step', x ='flipper_length_mm', hue = 'species')
plt.show()
```

```python
# 2. Stack bars on top of each other as shown below
sns.displot(penguins, x ='flipper_length_mm', hue = 'species', multiple ='stack')
plt.show()
```

```python
# 3. Use "dodge" to separate individual class . This will reduce the bar size, but ensures comparison by height
sns.displot(penguins, x= 'flipper_length_mm', hue ='species', multiple ='dodge')
plt.show()
```

```python
# Use displot() to facet on other variable.
sns.displot(penguins, x ='flipper_length_mm', col ='sex')
plt.show()
```

# Kernel Density Estimation (kde)

A histogram aims to approximate the underlying probability density function that generated the data by binning and counting observations.

Kernel density estimation presents a solution to the problem in a different way. Rather than using discrete bin, a KDE plot smooths the observation with a Gaussian kernel, producing a continuous estimate

```python
sns.displot(penguins, x = 'flipper_length_mm', kind ='kde')
plt.show()
```

```python
#Choosing the smoothing bandwidth
sns.displot(penguins, x ='flipper_length_mm', kind = 'kde', bw_adjust = .25)
plt.show()
```

Conditioning on other variables

```python
sns.displot(penguins, x = 'flipper_length_mm', hue = 'species', kind ='kde')
plt.show()
```

```python
# Layered KDE
sns.displot(penguins, x = 'flipper_length_mm', hue = 'species', kind = 'kde', multiple ='stack')
plt.show()
```

```python
#Plot both histogram and kde
sns.displot(penguins, x = 'flipper_length_mm', kde = True)
plt.show()
```

# Visualizing categorical data

Categorical plottings helps us to plot figures in which one of the main variables is categorical, i.e., divided into discrete groups
There are many ways to plot categorical plottings but the best way is to use the high-level API catplot() function

categorical plottings can be divided into 3 groups:
1. Categorical scatterplots
2. Categorical distribution plots
3. Categorical estimate plots

```python
#Categorical scatter plots
#Default is the stipplot, which applies some jitter to the observations on each category
sns.catplot(tips, x ='day', y ='total_bill')
plt.show()
```

The second approach adjusts the points along the categoricala axis using an algorithm that prevents them from overlapping

```python
sns.catplot(tips, x='day', y='total_bill',kind ='swarm')
plt.show()
```

```python
#Include the hue semantics
sns.catplot(data = tips, x ='day', y ='total_bill', hue = 'sex', kind ='swarm')
```
