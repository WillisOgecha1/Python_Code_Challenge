# MATPLOTLIB ADDING OF LABELS AND TITLE

```python
#Add titles and labels to your matplotlib graph

import numpy as np 
import matplotlib.pyplot as plt
pulse = np.array([80,85,90,95,100,105,110,115,120,125])
calorie = np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("Sports Watch Data")
plt.xlabel("Pulse rate")
plt.ylabel("Calorie Burnage")

plt.plot(pulse, calorie)
plt.show()
```

```python
#Set font properties for the labels and title
#Use the fontdict keyword argument
import numpy as np 
import matplotlib.pyplot as plt
pulse = np.array([80,85,90,95,100,105,110,115,120,125])
calorie = np.array([240,250,260,270,280,290,300,310,320,330])

font1= {'family':'serif', 'color':'blue', 'size':20}
font2= {'family':'serif', 'color':'darkred', 'size':15}

plt.title("Sports Watch Data", fontdict =font1)
plt.xlabel("Pulse rate", fontdict= font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(pulse, calorie, marker='D' ,mfc ='blue' ,mec ='blue',c = 'red', lw = 3.5)
plt.show()
```

```python
# Edit the location of the title,
#Posible positions inclued: center(which is the default location), left, right
# Use the loc keyword argument

import numpy as np
import matplotlib.pyplot as plt
pulse = np.array([80,85,90,95,100,105,110,115,120,125])
calorie = np.array([240, 250,260,270,280,290, 300, 310,320,330])

font1= {'family':'serif', 'color':'blue', 'size':20}
font2= {'family':'serif', 'color':'darkred', 'size':15}

plt.title("Sports Watch Data", fontdict =font1, loc = 'left')
plt.xlabel("Pulse rate", fontdict= font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.plot(pulse, calorie)
plt.show()

```

# MATPLOTLIB SUBLOTS:
With subplot() function, you can draw multiple plots in one figure

```python
#Draw two plots
import numpy as np
import  matplotlib.pyplot as plt
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
x = np.array([0,1,2,3])
y= np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

# SCATTER PLOT:

Scatter plot draws a dot for each observation.
You can use the scatter() function from pyplot to draw a scatter plot:

```python
#Example of a simple scatter plot
import  matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.show()
```

```python
#And color and size to the dots
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = ['red','brown','hotpink', 'lavender', 'indigo', 'blue', 'green', 'magenta','cyan','yellow','darkblue','grey','orange']

plt.scatter(x, y, c = colors)
plt.show()
```

```python
#Use color maps instead.
#Color maps is like a list of colors, with values from 0 to 100

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = [3,6,7,1,9,24,67,90,19,43,57,89,15,]

plt.scatter(x, y, c= colors, cmap = 'viridis')
plt.show()
```

```python
#Use color maps instead.
#Color maps is like a list of colors, with values from 0 to 100

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = [3,6,7,1,9,24,67,90,19,43,57,89,15,]

plt.scatter(x, y, c= colors, cmap = 'Set1')
plt.show()
```

```python
#change the size of the dots, using the 's' parameter

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = [3,6,7,1,9,24,67,90,19,43,57,89,15]
size = [250,11,49,58,87,99,321,272,15, 102,79,9,207]
plt.scatter(x, y, c= colors, cmap = 'Set1', s = size)
plt.show()
```

```python
#set the Alpha argument to inculcate transparency
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = [3,6,7,1,9,24,67,90,19,43,57,89,15]
size = [250,11,49,58,87,99,321,272,15, 102,79,9,207]
plt.scatter(x, y, c= colors, cmap = 'Set1', s = size, alpha =0.5)
plt.show()

```

# Plotting Bars.

With the pyplot, we can plot vertical and horizontal bars.
we use bar() and barh() functions, respectively.

```python
# Plot a simple  bar chart

import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

```python
#Plott horizontal bar
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.barh(x,y)
plt.show()
```

# BAR COLORS

```python
#Add colors to the bars using the keyword argument color
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y , color = 'Chocolate')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.barh(x,y , color = 'Khaki')
plt.show()
```

# Change the Bar width.
We use the width argument. 

For horizontal bars, we use height instead of width. The default size is 0.8

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.barh(x,y , color = 'Khaki', height = 0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array(['A', 'B', 'C','D'])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = 'Indigo', width = 0.5)

plt.show()
```

# MATPLOTLIB PIE CHART PLOTTING

```python
# PLot pie chart using the pie() function

import numpy as np
import matplotlib.pyplot as plt

y = np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
#Add labels to the pie chart

y = np.array([35,25,25,15])
mylabels = ['Watermelon', 'Bananas', 'Grapes', 'Mango'] 
plt.pie(y, labels = mylabels)
plt.show()
```

```python
# Change the start Angle
plt.pie(y, labels = mylabels, startangle = 180)
plt.show()
```

```python
# Explode: use it to make one or more wedge stand out.
myexplode = [0.2,0,0,0]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
# Add shadow to the pie chart by setting the shadow  parameter to True
myexplode = [0.2,0,0,0]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()

```

```python
# Set the color of each wedge with the colors parameter
mycolors = ['magenta', 'yellow','orange','g']
plt.pie(y, labels = mylabels, colors = mycolors, explode = myexplode, shadow = True)
plt.show()
```

```python
# Add legend 
mycolors = ['magenta', 'yellow','orange','g']
plt.pie(y, labels = mylabels, colors = mycolors, explode = myexplode, shadow = True)
plt.legend(title ='My fruits')
plt.show()
```
