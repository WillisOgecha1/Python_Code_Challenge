```python
#Function to reverse a lsit without using the built-in reverse method

def reverse_list(input_list):
    left = 0
    right = len(input_list)-1

    while left< right:
        input_list[left], input_list[right] = input_list[right], input_list[left]

        left +=1
        right -=1

    return input_list


my_list = [1,3,5,6,7,8,9]
result = reverse_list(my_list)
print(result)
```

```python
#Function to return the count of vowels;
def count_vowels(input_string):
    vowels = "aeiou"
    return sum(1 for char in input_string.lower() if char in vowels)

my_string = "My name is Jane"
print(count_vowels(my_string))
```

```python
#Defining the same function but with a longer approach

def count_vowels(input_string):
    vowels = "aeiou"
    count = 0

    for char in input_string.lower():
        if char in vowels:
            count +=1
    return count
my_string = "My name is Jane"
mylonger_string = "This is a brand new product, it requires much attention and time in order to be fully processed. Thank you"
print(count_vowels(my_string))
print(count_vowels(mylonger_string))    
```

```python
# Function to return a dictionary containing mean, max, min and std from a numpy 1D array

import numpy as np
def array_stats(arr):
    dict = {
        "mean" : np.mean(arr),
        "min" : np.min(arr),
        "max" : np.max(arr),
        "std" : np.std(arr)
    }
    return dict

arr = np.array([1,3,44,56,8,9,12,4,7])
print(array_stats(arr))
```

```python
# Function to check whether a phrase is a palindrome

def is_palindrome(input_string):
    #Clean the input string
    cleaned_string = input_string.replace(" ", "").lower()
    #Check for palindrome
    return cleaned_string == cleaned_string[::-1]
 #Test the function
string1= 'Race Car'
print(is_palindrome(string1))
string2 = "Hello"
string3 = "A man  a plan a canal  Panama"
print(is_palindrome(string2))
print(is_palindrome(string3))
```

```python
# A function to normalize a Numpy array input

def normalize_array(arr):
    mini = np.min(arr)
    maxi = np.max(arr)
    x = np.array(arr)
    normalised = (x-mini)/(maxi - mini)
    return normalised

arr = [2,4,5,7,8,10,11,14,24]
print(normalize_array(arr))
```

```python
# A function to generate fibonacci numbers

def fibonacci(n):
    if n<=0:
        return []
    if n<=1:
        return [0]
    fib_list = [0,1]

    while len(fib_list)<n:
        next_value = fib_list[-1] + fib_list[-2]
        fib_list.append(next_value)
    return fib_list    


#Example
print(fibonacci(7))
```
