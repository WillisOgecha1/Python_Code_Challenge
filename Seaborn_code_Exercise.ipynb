Challenge 1 — Relational Plot with Semantics
Using the tips dataset, create a scatter plot of total_bill vs tip, where:

Color encodes smoker, 
Style encodes smoker, 

Size encodes size, 
Use a figure-level function

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_theme(style ='darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#Plot the figure
sns.relplot(data = tips, x ='total_bill', y ='tip', hue ='smoker', style = 'smoker', size = 'size')
plt.show()
```

Challenge 2 — Faceted Line Plot

Using the fmri dataset, plot signal over timepoint as a line plot, faceted by event across columns and region across rows. Disable the confidence interval.

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(fmri, kind ='line', x ='timepoint', y = 'signal', col ='event', row ='region', estimator = None)
plt.show()
```

Challenge 3 — Distribution Comparison

Using the penguins dataset, plot the distribution of flipper_length_mm for each species using KDE, all on the same axes, with each species stacked. Use a figure-level function.

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
sns.displot(penguins, x='flipper_length_mm', kind = 'kde', hue = 'species', multiple = 'stack')
plt.show()
```

Challenge 4 — Categorical Plot

Using the titanic dataset, create a violin plot of age grouped by class on the x-axis, colored by sex, with the violins split.

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.catplot(titanic, x= 'class', y = 'age', hue = 'sex',kind='violin', split = True)
plt.show()
```
