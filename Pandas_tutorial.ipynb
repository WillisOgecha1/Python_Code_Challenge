```python
import pandas as pd

mydataset = {
    'cars':["BMW", "Volvo", "Toyota", "Ford"],
    'passings':[3,2,7,5]
}

myvar = pd.DataFrame(mydataset)
print(myvar)
```

```python
#Creating pandas series.
#1. From a list

a = [1,7,3,4,6,9]
myvar = pd.Series(a)
print(myvar)
```

```python
#Use label to access pandas series' data item
print(myvar[0])
```

```python
#We can also use index argument to assign the labels as below
mayvar = pd.Series(a, index =["u","v","w", "x","y","z"])
print(mayvar)
```
