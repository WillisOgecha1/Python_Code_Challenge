```python
import pandas as pd
data = {
    "Player": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Goals": [5, 12, 7, None, 9],
    "Assists": [3, None, 6, 8, 2],
    "Minutes": [90, 85, 120, 95, 200]
}
df = pd.DataFrame(data)

df.fillna({'Goals' :df['Goals'].median()}, inplace =True)
df.fillna({'Assists':df['Assists'].mode()}, inplace =True)

#Find and print any rows where Minutes is greater than 100 — these are outliers.
for x in df.index:
    if df.loc[x, 'Minutes']>100:
        print(df.loc[x])
#Replace any Minutes value above 100 with 100.
for x in df.index:
    if df.loc[x,'Minutes']>100:
        df.loc[x, 'Minutes']=100
#Print the DataFrame
print(df.to_string())
        
```

```python
import pandas as pd

data = {
    "Product": ["Laptop", "Phone", "Tablet", "Monitor", "Laptop"],
    "Price": [999, None, 499, 299, 999],
    "Rating": [4.5, 3.8, None, 4.1, 4.5],
    "Sales": [150, 200, 80, 500, 150]
}

df = pd.DataFrame(data)
df.fillna({'Price': df['Price'].mean()}, inplace = True)
df.fillna({'Rating': df['Rating'].median()}, inplace = True)
df.drop_duplicates(inplace = True)

correlation = df['Price'].corr(df['Sales'])
print(f"The correlation between Price and Sales is:{correlation}")

print(df.to_string())
```
