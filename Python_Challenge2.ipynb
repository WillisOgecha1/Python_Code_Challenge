```python
#CORRECT SOLUTION
def flatten_list(lst):
    flat_list =[]
    def recursive_flatten(items):
        for item in items:
            if isinstance(item, list):
                recursive_flatten(item)
            else: 
                flat_list.append(item)
    recursive_flatten(lst)

    seen = list(dict.fromkeys(flat_list))
    return sorted(seen)

nested = [3, [1, 2, [2, 4]], [3, [5, 1]], 6]

flatten_list(nested)
    
```

```python
from collections import deque
def sliding_window_max(nums, k):
    if not nums:
        return []
    result = []
    dq = deque()
    for i in range(len(nums)):
        if dq and dq[0]<i-k+1:
            dq.popleft()
        while dq and nums[dq[-1]]<nums[i]:
            dq.pop()
        dq.append(i)
        if i>= k-1:
            result.append(nums[dq[0]])
    return result

nums = [1, 3, -1, -3, 5, 3, 6, 7]
k = 3
print(sliding_window_max(nums, k))
```

```python
from collections import OrderedDict
class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = OrderedDict()
    def get(self, key):
        if key not in self.cache:
            return -1
        self.cache.move_to_end(key)
        return self.cache[key]
    def put(self, key, value):
        if key in self.cache:
            self.cache.move_to_end(key)
        self.cache[key] = value
        if len(self.cache) > self.capacity:
          self.cache.popitem(last = False)

cache = LRUCache(capacity=3)

cache.put(1, "one")
cache.put(2, "two")
cache.put(3, "three")

cache.get(1) 
```

```python
def generate_parentheses(n):
    result = []
    
    def backtrack(current_string, left_count, right_count):
        # Base case: If the string is full length, we've found a valid combo
        if len(current_string) == 2 * n:
            result.append(current_string)
            return
        
        # Rule 1: We can always add a '(' if we haven't used all 'n' of them
        if left_count < n:
            backtrack(current_string + "(", left_count + 1, right_count)
            
        # Rule 2: We can only add a ')' if there are more '(' than ')' 
        # currently in the string (this prevents closing before opening)
        if right_count < left_count:
            backtrack(current_string + ")", left_count, right_count + 1)

    backtrack("", 0, 0)
    return result

# Example for n = 3
print(generate_parentheses(3))
# Output: ["((()))", "(()())", "(())()", "()(())", "()()()"]
```
